```python

import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt # to plot the accuracy graph
```

```python
# Task 1: in this task we will use SVM for multi-class classification
# 1.1: Read the Iris dataset with the following columns 'sepal length','sepal width','petal length','petal width','target'
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
df.head()
```

```python
# 1.2: extract the inputs into X and target into y
X = df.loc[:,['sepal length','sepal width','petal length','petal width']]
y = df.loc[:,['target']]
```

```python
# 1.3: Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)
```

```python
# 1.4: Create the SVM Model, create three models with three different kernels 'poly', 'sigmoid', and 'linear'.
# for more information about the kernels, visit
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

model_a = svm.SVC(kernel='poly', random_state=42)
model_b = svm.SVC(kernel='sigmoid', random_state=42)
model_c = svm.SVC(kernel='linear', random_state=42)
```

```python

# 1.5: train the three model using the training data
model_a.fit(X_train, y_train)
model_b.fit(X_train, y_train)
model_c.fit(X_train, y_train)


# 1.6: test the three model using the testing data
pred_a = model_a.predict(X_test)
pred_b = model_b.predict(X_test)
pred_c = model_c.predict(X_test)
```

```python
# 1.7: print the results as accuracy for each model
print("Poly Training Accuracy:", model_a.score(X_train, y_train))
print("Poly Testing Accuracy:", model_a.score(X_test, y_test))
print("Sigmoid Training Accuracy:", model_b.score(X_train, y_train))
print("Sigmoid Testing Accuracy:", model_b.score(X_test, y_test))
print("Linear Training Accuracy:", model_c.score(X_train, y_train))
print("Linear Testing Accuracy:", model_c.score(X_test, y_test))
```

```python
# Additional: Tuning the parameters for the machine learning model using Grigsearchcv()
#this is used to try different parameters values in order to tune the model by finding the best set of parameters
# we call this optimization, more on this:
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html

from sklearn.model_selection import GridSearchCV
# we will use the optimazer on the SVM model
Model_SVM = svm.SVC()

#define the parameters for the SVM that we want to optimise
hyperparameter_space = [{'C':[0.1, 1, 10, 100],'kernel':['rbf','sigmoid','linear']}]

# create our optimiser using the set of parametrs
optimizer = GridSearchCV(Model_SVM, param_grid=hyperparameter_space,
                  scoring="accuracy", cv=2, return_train_score=True)

# train the model using the optimizer (train with tuning to find the best set of parameters)
optimizer.fit(X_train, y_train)

# print the results
print("Optimal hyperparameter combination:", optimizer.best_params_)
print()
print("Mean cross-validated training accuracy score:",optimizer.best_score_)

# use the best parameters set to train the module
optimizer.best_estimator_.fit(X_train, y_train)
# print the train results
print("Train accuracy:", optimizer.best_estimator_.score(X_train, y_train))

# use the trained model to test
y_pred = optimizer.best_estimator_.predict(X_test) # Predictions

# print the test results

print("Test accuracy:", metrics.accuracy_score(y_test, y_pred))
```
